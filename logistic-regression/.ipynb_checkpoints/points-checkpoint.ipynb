```python
import matplotlib.pyplot as plt
import numpy as np
import random

from math import log, exp
```

# Generating and classifying data

```python
def random_on_rect(span_x, span_y, total):
    return np.array([[random.uniform(*span_x), random.uniform(*span_y)] for _ in range(total)])
```

```python
def get_classification(points, slope, constant):
    def classify_point(point):
        x, y = point
        if y >= slope * x + constant:
            return 1
        else:
            return 0
    return np.array([classify_point(point) for point in points])

def split_into_classes(points, classifications):
    above, below = [], []
    for point, classification in zip(points, classifications):
        if classification == 1:
            above.append(point)
        elif classification == 0:
            below.append(point)
    return above, below
```

```python
class Visualiser:
    def show_classified(points, slope, constant):
        above, below = split_into_classes(points, get_classification(points, 1, 0))
        plt.scatter(*zip(*above))
        plt.scatter(*zip(*below))
        plt.show()
```

```python
points = random_on_rect((-1, 1), (-1, 1), 1000)
Visualiser.show_classified(points, 2, 0)
```

## Helper functions
Because I wanted to try different activation functions there is a single activation function. If you want to switch the sigmoid function to anything else you can simply change the return statement and nothing else.

```python
def sigmoid(x):
    return 1 / ( 1 + exp(-x) )

def activation(x):
    return sigmoid(x)
```

## Initializing data

```python
def initialize_weights(data):
    weights = np.zeros(((data.shape)[1]+1,1)) # Initialize weights as 0
    data = np.c_[np.ones(((data.shape)[0], 1)), data] # I add a bias to the data vectors
    return weights, data
```

I'll be using this function for calculating loss:
